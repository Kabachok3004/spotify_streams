```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
from sklearn.preprocessing import PolynomialFeatures

data_path = "/content/spotify-2023.xlsx"

# С кодировкой файла возникли проблемы так что пришлось конвертировать его через excel(
print(os.path.getsize(data_path))

df = pd.read_excel(data_path)

df.head()
```

```python
df = df.dropna()
df.describe()
```

```python
import seaborn as sns

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

df['bpm'].plot(kind='hist', bins=20, ax=axes[0], title='BPM Distribution')
sns.countplot(x='key', data=df, ax=axes[1], order=df['key'].value_counts().index, palette='viridis')
sns.countplot(x='mode', data=df, ax=axes[2], order=df['mode'].value_counts().index, palette='viridis')

plt.tight_layout()
plt.show()

correlation_matrix = df[['danceability_%', 'valence_%', 'energy_%', 'acousticness_%',
                         'instrumentalness_%', 'liveness_%', 'speechiness_%','artist_count', 'in_spotify_playlists', 'in_spotify_charts', 'streams']].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

Видно что параметры 'in_spotify_playlists' и 'in_spotify_charts' значительно влияют на количество прослушиваний, так что будет логичным взять модель обученную только на этих параметрах за основу и сравнивать с ней модель учитывающую все параметры

```python
data_limited = df[['in_spotify_playlists', 'in_spotify_charts', 'streams']]
X, y = data_limited.drop('streams', axis=1), data_limited['streams']
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size = 0.25)
```

```python
poly_mod = PolynomialFeatures(degree=2)
poly_mod.fit(Xtrain)
Xtrain_poly = poly_mod.fit_transform(Xtrain)
Xtest_poly = poly_mod.transform(Xtest)
```

```python
linear_reg = LinearRegression()
linear_reg.fit(Xtrain, ytrain)
ytest_pred = linear_reg.predict(Xtest)
```

```python
linear_reg_poly = LinearRegression()
linear_reg_poly.fit(Xtrain_poly, ytrain)
ytest_pred_poly = linear_reg_poly.predict(Xtest_poly)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

mspe_test = ((ytest - ytest_pred) / ytest)**2
mape_test = abs((ytest - ytest_pred) / ytest)
mspe_test_poly = ((ytest - ytest_pred_poly) / ytest)**2
mape_test_poly = abs((ytest - ytest_pred_poly) / ytest)

print("Метрики для линейной регрессии:")
print("MSE:", mean_squared_error(ytest, ytest_pred))
print("MAE:", mean_absolute_error(ytest, ytest_pred))
print("MSPE:", mspe_test.mean())
print("MAPE:", mape_test.mean())
print("Метрики для линейной регрессии с полиномиальными признаками:")
print("MSE:", mean_squared_error(ytest, ytest_pred_poly))
print("MAE:", mean_absolute_error(ytest, ytest_pred_poly))
print("MSPE:", mspe_test_poly.mean())
print("MAPE:", mape_test_poly.mean())
```

Для этих двух параметров поленемиальные признаки бесполезны

Так-же мы видим что оценка прослушиваний оочень грубая, попробуем хоть немного улучшить результат за счет других признаков

Все признаки скалярные кроме столбцов задающих тональность.
Видно что нет ни одной строки с нотой ДО, думаю это неточность датасета и пропущенные значения в этом столбе как раз таки имеют значение С.
При применении OneHot encoding это значение все равно можно будет удалить из-за линейной зависимости одного столбца, но ради приличия дополним датасет

Логично что текстовые значения вроде названия трека и имени артиста нужно удалить, но в будущем можно будет попробовать реализовать и эти данные за счет их эмбеддингов

```python
from sklearn.preprocessing import OneHotEncoder

df['key'].fillna('C', inplace=True)
df = df.drop(['track_name','artist(s)_name'], axis=1)
df['mode'] = df['mode'].apply(lambda x: 1 if x == 'Major' else 0)
df_without_tonal = df.drop(['key','mode'], axis=1)
df_encoded = pd.get_dummies(df, columns=['key'], prefix='key')
df_encoded = df_encoded.drop('key_A#', axis=1)
df_encoded.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X, y = df_encoded.drop('streams', axis=1), df_encoded['streams']

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size = 0.25)

X_train_scaled = scaler.fit_transform(Xtrain)

X_test_scaled = scaler.transform(Xtest)
```

```python
poly_mod2 = PolynomialFeatures(degree=2)

X_TrSP = poly_mod2.fit_transform(X_train_scaled)
X_TeSP = poly_mod2.transform(X_test_scaled)
```

```python
linear_reg_poly = LinearRegression()

linear_reg_poly.fit(X_TrSP, ytrain)

ytest_pred = linear_reg_poly.predict(X_TeSP)

mse_test = mean_squared_error(ytest, ytest_pred)
mae_test = mean_absolute_error(ytest, ytest_pred)
mspe_test = ((ytest - ytest_pred) / ytest)**2
mape_test = abs((ytest - ytest_pred) / ytest)

print("Метрики для линейной регрессии на тестовом наборе:")
print("MSE:", mse_test)
print("MAE:", mae_test)
print("MSPE:", mspe_test.mean())
print("MAPE:", mape_test.mean())
```

Кажется модель переобучилась, возможно ее запутали категрии с тональностями, попробуем обойтись без них

```python
X = df_without_tonal.drop('streams', axis=1)
y = df_without_tonal['streams']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

poly_mod = PolynomialFeatures(degree=2)

X_train_poly = poly_mod.fit_transform(X_train)
X_test_poly = poly_mod.transform(X_test)

linear_reg_poly = LinearRegression()

linear_reg_poly.fit(X_train_poly, y_train)

y_test_pred = linear_reg_poly.predict(X_test_poly)

mse_test = mean_squared_error(y_test, y_test_pred)
mae_test = mean_absolute_error(y_test, y_test_pred)
mspe_test = ((y_test - y_test_pred) / y_test)**2
mape_test = abs((y_test - y_test_pred) / y_test)

print("Метрики для линейной регрессии на тестовом наборе:")
print("MSE:", mse_test)
print("MAE:", mae_test)
print("MSPE:", mspe_test.mean())
print("MAPE:", mape_test.mean())
```

Все же в конце концов модель переобучается, похоже задача предсказания прослушиваний выходит за рамки применения линейной регрессии, но все же можно делать групые оценки на основании использовавшихся в начале двух параметров
